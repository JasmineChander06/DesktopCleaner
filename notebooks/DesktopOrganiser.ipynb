```python
#now to get all the files paths the same we are going to create list and save it under each of the extension folder 

# define desktop path 
desktop_path = os.path.expanduser("~/OneDrive/Desktop")
print(desktop_path)

# Creating a dictionary to save all the file paths by file extension 
files_by_ext={}

#Loop through all the files 
for filename in os.listdir(desktop_path):
       file_path= os.path.join(desktop_path,filename)

    #Now we'll check if it is a file or not because of our logic if it is a folder it would be ignored 
       if os.path.isfile(file_path):
        file_extension=os.path.splitext(filename)[1]
        # adding file extension and file path to the dictionary
        if file_extension not in files_by_ext:
            files_by_ext[file_extension]=[]
        files_by_ext[file_extension].append(file_path)    
        
        
```

```python
# Create a folder for each file and moving the respective folder 
for file_extension,file_paths in files_by_ext.items():
    folder_name=f"{file_extension[1:].upper()} Files"
    folder_path=os.path.join(desktop_path,folder_name)
    # Create the folder if it doesnt exist 
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
    # moving the files to the respective folders 
    for file_path in file_paths:
        shutil.move(file_path,folder_path)
    
```
